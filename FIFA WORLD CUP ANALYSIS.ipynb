# Feature Engineering

```python
# Example Feature Engineering: Creating historical performance indicators

# Calculate win percentages for each team
team_wins = world_cup_matches['Winners'].value_counts()
team_matches_played = (
    world_cup_matches['Home Team Name'].value_counts() + world_cup_matches['Away Team Name'].value_counts()
)
win_percentages = (team_wins / team_matches_played).fillna(0)

# Create a new feature 'Win Percentage' for each team in the World Cups dataset
world_cups['Win Percentage'] = world_cups['Winner'].map(win_percentages)

# You can add more features relevant to your analysis

# Display the updated World Cups dataset with new features
print(world_cups.head())
```

```python
def Average_Attendance_by_year(data):
    att1 = data.groupby("Year")["Attendance"].sum().reset_index()
    att1["Year"] = att1["Year"].astype(int)
    plt.figure(figsize=(12, 7))
    ax = sns.pointplot(data=att1, x="Year", y="Attendance", color="w")
    ax.set_facecolor("k")
    plt.xlabel("Year", fontsize=20)
    plt.ylabel("Attendance", fontsize=20)
    plt.grid(True, color="grey", alpha=0.3)
    plt.title("Average Attendance In Every World Cup year", fontsize=25, color="green")
    plt.show()

# Call the function with the appropriate data (e.g., world_cup_matches)
Average_Attendance_by_year(world_cup_matches)
```

```python
def label(world_cup_matches):
    if world_cup_matches["Home Team Goals"] > world_cup_matches["Away Team Goals"]:
        return "Home team win"
    if world_cup_matches["Away Team Goals"] > world_cup_matches["Home Team Goals"]:
        return "Away team win"
    if world_cup_matches["Home Team Goals"] == world_cup_matches["Away Team Goals"]:
        return "DRAW"

world_cup_matches["outcome"] = world_cup_matches.apply(lambda world_cup_matches:label(world_cup_matches),axis=1)
plt.figure(figsize=(9,9))
world_cup_matches["outcome"].value_counts().plot.pie(autopct="%1.0f%%",fontsize =14,
                                           colors = sns.color_palette("husl"),
                                           wedgeprops={"linewidth":2,"edgecolor":"white"},
                                           shadow=True)
circ = plt.Circle((0,0),.7,color="white")
plt.gca().add_artist(circ)
plt.title("Match outcomes based on home and away teams" , fontsize = 25 , color = "red")
plt.show()
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table of Home Team Win vs. Not Home Team Win
contingency_table = pd.crosstab(world_cup_matches['Winners'], world_cup_matches['Home Team Name'] == world_cup_matches['Winners'])

# Perform a chi-squared test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

if p_value < 0.05:
    print(f'The difference in win percentages is statistically significant (p-value: {p_value:.4f})')
else:
    print(f'The difference in win percentages is not statistically significant (p-value: {p_value:.4f})')
```

```python
contingency_table = pd.crosstab(world_cup_matches['Winners'], world_cup_matches['Away Team Name'] == world_cup_matches['Winners'])

# Perform a chi-squared test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

if p_value < 0.05:
    print(f'The difference in win percentages is statistically significant (p-value: {p_value:.4f})')
else:
    print(f'The difference in win percentages is not statistically significant (p-value: {p_value:.4f})')
```
